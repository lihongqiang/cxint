```python
wn_des = open("./label/wordnet_openimage_description.tsv", "r")
wn_des_dict = {}
for line in wn_des:
    wnID,wnlox,des = line.split('\t')
    des = des.strip('\n')
    wn_des_dict[wnID] = wnlox+":"+des
    print wnID,wn_des_dict[wnID]
wn_des.close()
```

```python
import xlrd
import codecs
data_human = xlrd.open_workbook("./label/data_humman.xlsx", "r")
data_human_merg = codecs.open("./label/data_humman_des.tsv", "wb", "utf8")
booksheet = data_human.sheet_by_name('wordnet_openimage_label - 2016.')
num = 0
for row in range(booksheet.nrows):
    labelID = booksheet.cell(row, 0).value
    cnt = booksheet.cell(row, 1).value
    labelName = booksheet.cell(row, 2).value
    wnID = booksheet.cell(row, 3).value.strip('\n\t')
    if wnID not in wn_des_dict:
        data_human_merg.write(labelID + "\t" + str(int(cnt)) + "\t" + labelName + "\t" + wnID + "\t" + ''+ "\n")
        num += 1
    else:
        data_human_merg.write(labelID + "\t" + str(int(cnt)) + "\t" + labelName + "\t" + wnID + "\t" + wn_des_dict[wnID] + "\n")
data_human_merg.close()
print num
```
