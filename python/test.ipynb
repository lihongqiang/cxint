```python
from nltk.corpus import wordnet as wn
des_file = open("./data/wordnet_openimage_description.tsv", "r")
syn_words = {}
syn_des = {}
flag = 0
total = 0
for line in des_file:
    wnid = line.split('\t')[0]
    syn_des [wnid] = line.split('\t')[2].strip('\n')
    total += 1
    p = wnid[0]
    off = wnid[1:]
    syn = wn._synset_from_pos_and_offset(p,int(off))
    words = sorted(syn.lemma_names())
    words = '_'.join(words)
    if words not in syn_words:
        syn_words[words] = wnid
    else:
        flag += 1
        print words,wnid,syn_words[words]
        print wnid + "\t" + syn_des[wnid]
        print syn_words[words] + "\t" + syn_des[syn_words[words]] +'\n'
if flag:
    print ''
else:
    print ''
print str(flag) + '/' + str(total)
```
