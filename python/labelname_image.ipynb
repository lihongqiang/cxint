下载label_image_dict对应的所有图片

```python
#-*-coding:utf-8-*-
import sys
import os
import requests
import threading
import time
imageID_dir = u'./images/'

label_image_list = label_image_dict.items()
label_num = len(label_image_list)
print label_num
thread_count = 1
thread = []
interval =  label_num/ 10
start = 0
end = interval
def downloadImg(label_image_list,dir_path,start,end):
    
    print('thread %s is running...' % threading.current_thread().name)
    
    thread_name = threading.current_thread().name
    img_count = 0
    
    for label,imageID_list in label_image_list[start:end]:
        
        label_dir = dir_path + label_name[label].replace(' ', '_').decode('UTF-8')
        
        if not os.path.exists(label_dir):
            print 'dir ' + label_dir
            os.makedirs(label_dir)
            
        print "%s正在下载%d/%d个标签%s %s" % (thread_name,img_count,end - start,label,label_name[label])
        for imageID in imageID_list:
            img_url = imageID_url_dict[imageID]
            image_path = label_dir+'/'+imageID+'.'+img_url.split('.')[-1]
            if not os.path.exists(image_path):
                img_content = requests.get(img_url)
                if img_content.status_code == 200:
                    print image_path, img_url
                    with open(image_path,'wb') as f:
                        f.write(img_content.content)
                        f.close()
                else:
                    print "status "+img_content.status_code
        img_count = img_count + 1

    print('thread %s ended.' % threading.current_thread().name)
    
    
while end <= label_num:
    t = threading.Thread(target=downloadImg,args=(label_image_list,imageID_dir,start,end),name="Thread" + str(thread_count))
    thread.append(t)
    start = end
    end = end + interval
    thread_count = thread_count + 1
    
for t in thread:
    t.start()
for t in thread:
    t.join()
        
```
